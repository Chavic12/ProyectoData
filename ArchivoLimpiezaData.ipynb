# Similitud con la tecnica del coseno

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Ejemplo de DataFrames
dataframe1 = data_frame

dataframe2 = df


def encontrar_similitud(carrera, materia):
    materias_dataframe2 = dataframe2[dataframe2['nombreCarrera'] == carrera]['nombreMateria']
    
    if not materias_dataframe2.empty:
        # Filtrar valores nulos en las materias
        materias_dataframe2 = materias_dataframe2.dropna()
        
        if not materias_dataframe2.empty:
            # Calcular similitud de coseno solo si hay materias no nulas en el DataFrame 2
            vectorizer = CountVectorizer().fit_transform([materia] + list(materias_dataframe2))
            vectors = vectorizer.toarray()
            csim = cosine_similarity(vectors)
    
            # Encontrar el índice de la similitud máxima (excluyendo la primera fila, que es la materia del dataframe1)
            max_similarity_index = csim[0, 1:].argmax()
    
            # Obtener la materia más similar
            materia_similar = materias_dataframe2.iloc[max_similarity_index]
        else:
            materia_similar = None
    else:
        materia_similar = None
    
    return materia_similar

# Aplicar la función a cada fila del dataframe1
dataframe1['MATERIA_SIMILAR'] = dataframe1.apply(lambda row: encontrar_similitud(row['MALLA_NOMBRE_LARGO'], row['columnaConcatenada']), axis=1)

dataframe1
```

# Simitud con la libreria fuzzywuzzy

```python
from fuzzywuzzy import fuzz
dataframe1 = data_frame

dataframe2 = df
def encontrar_similitud(carrera, materia):
    materias_dataframe2 = dataframe2[dataframe2['nombreCarrera'] == carrera]['nombreMateria']
    
    if not materias_dataframe2.empty:
        max_similarity = 0
        materia_similar = None
        
        for materia2 in materias_dataframe2:
            # Convertir los valores a cadenas antes de calcular la similitud
            materia_str = str(materia)
            materia2_str = str(materia2)
            
            similarity_ratio = fuzz.ratio(materia_str, materia2_str)
            if similarity_ratio > max_similarity:
                max_similarity = similarity_ratio
                materia_similar = materia2
        
        return materia_similar
    
    return None

# Aplicar la función a cada fila del dataframe1
dataframe1['MATERIA_SIMILAR'] = dataframe1.apply(lambda row: encontrar_similitud(row['MALLA_NOMBRE_LARGO'], row['columnaConcatenada']), axis=1)

# Mostrar el resultado
dataframe1[dataframe1["MALLA_NOMBRE_LARGO"] == "COMPUTACION"]
```
